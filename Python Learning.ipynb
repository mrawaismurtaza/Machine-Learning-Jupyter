```python
import numpy as np
import cv2
```

```python
# Video Capture and Window Sizes
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    width = int(cap.get(3))
    height = int(cap.get(4))
    
    image = np.zeros(frame.shape, np.uint8)

    smaller_frame = cv2.resize(frame, (0,0), fx=0.5, fy=0.5)
    image[:height//2, :width//2] = cv2.rotate(smaller_frame, cv2.ROTATE_180)
    image[height//2:, :width//2] = smaller_frame
    image[:height//2, width//2:] = cv2.rotate(smaller_frame, cv2.ROTATE_180)
    image[height//2:, width//2:] = smaller_frame
    
    cv2.imshow('frame', image)
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
# Line, Circle, Images and Rectangle DRAW in Video Frame
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    width = int(cap.get(3))
    height = int(cap.get(4))

    image = cv2.line(frame, (0,0), (width, height), (255,0,0), 5)
    image = cv2.line(image, (0, height), (width, 0), (0,255,0), 2)
    image = cv2.rectangle(image, (100, 100), (300,200), (0,0,255), -1)
    image = cv2.circle(image, (300,300), 60, (255,255,0), 3) 

    font = cv2.FONT_HERSHEY_SIMPLEX
    image = cv2.putText(image, 'Hello World!', (380, height - 20), font, 3, (255,0,0), 3) #position, font style, text size, color, width
    
    cv2.imshow('frame', image)
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
# Colors and Color Detection
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    width = int(cap.get(3))
    height = int(cap.get(4))

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower_blue = np.array([90, 50, 50])
    upper_blue = np.array([130, 255, 255])

    mask = cv2.inRange(hsv, lower_blue, upper_blue)

    result = cv2.bitwise_and(frame, frame, mask=mask)
    
    cv2.imshow('frame', result)
    cv2.imshow('mask', mask)
    cv2.imshow('hsv', hsv)
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
